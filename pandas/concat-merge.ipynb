# Concat

```python
import pandas as pd
```

```python
jan_df = pd.DataFrame({
    'player1': [10, 20, 30],
    'player2': [30, 20, 50],
    'player3': [80, 40, 30]
}, index=[0,1,2])
jan_df
```

```python
feb_df = pd.DataFrame({
    'player1': [40, 20, 30],
    'player2': [90, 50, 10],
    'player3': [80, 10, 80]
}, index=[3,4,5])
feb_df
```

```python
mar_df = pd.DataFrame({
    'player1': [90, 20, 60],
    'player2': [30, 80, 10],
    'player3': [50, 80, 80]
}, index=[6,7,8])
mar_df
```

```python
# concat data frames row wise
pd.concat([jan_df, feb_df, mar_df])
```

```python
# concat data frames column wise
pd.concat([jan_df, feb_df, mar_df], axis=1)
```

# Merge

```python
# customer data frame
customer_df = pd.DataFrame({
    'customer_id': ['c1', 'c2', 'c3'],
    'name': ['John Wick', 'Rocky', 'Tintin']
})
customer_df
```

```python
customer_order_map_df = pd.DataFrame({
    'customer_id': ['c1', 'c2', 'c3', 'c1', 'c1', 'c3'],
    'order_id': ['o1', 'o2', 'o3', 'o4', 'o5', 'o6']
})
customer_order_map_df
```

```python
order_df = pd.DataFrame({
    'price': [100, 200, 300, 100, 300, 500],
    'order_id': ['o1', 'o2', 'o3', 'o4', 'o5', 'o6']
})
order_df
```

```python
customer_order_df = pd.merge(customer_df, customer_order_map_df, how='inner', on='customer_id')
customer_order_df
```

```python
customer_order_price_df = pd.merge(customer_order_df, order_df, how='inner', on='order_id')
customer_order_price_df
```

```python
customer_order_price_df.groupby('customer_id').sum()
```
